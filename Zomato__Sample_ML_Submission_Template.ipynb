## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# RELOAD DATASETS TO ENSURE CLEAN STATE FOR 'Time' COLUMN
df1=pd.read_csv('Zomato Restaurant names and Metadata.csv')
df2=pd.read_csv('Zomato Restaurant reviews.csv')
merge_data=pd.merge(df1,df2,left_on='Restaurant',right_on='Restaurant',how='inner')

# Also convert Cost to numeric (doing this early to ensure consistent numerical type)
merge_data['Cost'] = merge_data['Cost'].astype(str).str.replace(',', '').astype(float)

# **Fix for Rating column**:
# Replace "No rating" string with NaN first, then convert to numeric (float).
merge_data['Rating'] = merge_data['Rating'].replace('No rating', np.nan)
merge_data['Rating'] = pd.to_numeric(merge_data['Rating'], errors='coerce') # Convert to numeric, coercing errors to NaN
# Fill remaining NaNs with a numerical value, e.g., 0 for missing ratings, then convert to int.
merge_data['Rating'] = merge_data['Rating'].fillna(0).astype(int)

# **Fix for Time column (moved before drop_duplicates)**:
# Ensure 'Time' is string type first.
merge_data['Time_datetime'] = pd.to_datetime(merge_data['Time'].astype(str), format='%m/%d/%Y %H:%M', errors='coerce')

# Convert datetime objects to Unix timestamps (numeric) for mean calculation.
numeric_time = merge_data['Time_datetime'].apply(lambda x: x.timestamp() if pd.notna(x) else np.nan)

# Calculate mean before filling NaNs
mean_numeric_time = numeric_time.mean()

# Fill missing values in the numerical timestamp series with the calculated mean.
merge_data['Time'] = numeric_time.fillna(mean_numeric_time)

# Drop the intermediate datetime column
merge_data = merge_data.drop(columns=['Time_datetime'])

# 1. Remove duplicate rows (now applied AFTER Time and Rating processing)
print(f"Time nulls BEFORE drop_duplicates: {merge_data['Time'].isnull().sum()}")
merge_data=merge_data.drop_duplicates()
print(f"Time nulls AFTER drop_duplicates: {merge_data['Time'].isnull().sum()}")

#2. Filling missing collections with "not available"
merge_data['Collections']=merge_data['Collections'].fillna("not available")

#3. Filling missing timings with "not available"
merge_data['Timings']=merge_data['Timings'].fillna("not available")

#4. Filling Reviewer with "NIL"
merge_data['Reviewer']=merge_data['Reviewer'].fillna("NIL")

#5. Filling Review with "No reviews"
merge_data['Review']=merge_data['Review'].fillna("No review")

#6. Filling Metadata with "No data"
merge_data['Metadata']=merge_data['Metadata'].fillna("No data")

# Extract city from Links and add it as a new column
merge_data['city'] = merge_data['Links'].str.split('/').str[3]

print("\nFinal information:")
merge_data.info()
print("\nFirst few rows of processed data:")
display(merge_data.head())
````

````python
# Chart - 7 visualization code

# Split cuisines and explode to get individual cuisine entries
cuisine_ratings = merge_data[['Cuisines', 'Rating']].copy()
cuisine_ratings['Cuisine'] = cuisine_ratings['Cuisines'].astype(str).str.split(', ')
cuisine_ratings = cuisine_ratings.explode('Cuisine')

# Ensure 'Rating' column is numeric before calculating mean
cuisine_ratings['Rating'] = pd.to_numeric(cuisine_ratings['Rating'], errors='coerce')

# Calculate average rating for each cuisine
average_cuisine_rating = cuisine_ratings.groupby('Cuisine')['Rating'].mean().sort_values(ascending=False)

# Get the top 10 rated cuisines (excluding 'not available' if it appeared due to NaN filling)
top_10_rated_cuisines = average_cuisine_rating[average_cuisine_rating.index != 'not available'].head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_rated_cuisines.values, y=top_10_rated_cuisines.index, palette='magma')
plt.title('Top 10 Cuisines by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Cuisine', fontsize=12)
plt.tight_layout()
plt.show()
````

````python
import logging
import optuna
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# 1. Enable logging so you see progress every few seconds
optuna.logging.set_verbosity(optuna.logging.INFO);

# 2. Use a subset (e.g., 1000 rows) for lightning-fast tuning
X_train_sub = X_train # Assuming X_train is already a suitable type for .sample or directly for subsetting
if hasattr(X_train, 'sample') and len(X_train) > 1000: # Reduced subset size
    X_train_sub = X_train.sample(n=1000, random_state=42)
    y_train_sub = y_train.loc[X_train_sub.index]
else:
    X_train_sub = X_train
    y_train_sub = y_train

def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 10, 50),
        'max_depth': trial.suggest_int('max_depth', 2, 5),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 5),
        'bootstrap': True,
        'n_jobs': -1
    }

    model = RandomForestRegressor(**params, random_state=42);

    score = cross_val_score(model, X_train_sub, y_train_sub, cv=3,
                            scoring='neg_mean_squared_error', n_jobs=-1).mean();
    return score

# 3. Create and run the study
study = optuna.create_study(direction='maximize');
study.optimize(objective, n_trials=5); # Reduced number of trials

# 4. Train the FINAL model on the FULL data using the best params
print(f"Best Params found: {study.best_params}")
best_model = RandomForestRegressor(**study.best_params, random_state=42, n_jobs=-1);
best_model.fit(X_train, y_train);

# Calculate evaluation metrics for the best model
cv_score = cross_val_score(best_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error');
cv_mse = abs(cv_score.mean());

y_pred = best_model.predict(X_test);
mse = mean_squared_error(y_test, y_pred);
mae = mean_absolute_error(y_test, y_pred);
r2 = r2_score(y_test, y_pred);

print(f"\nCross-validation MSE for tuned model: {cv_mse:.4f}")
print(f"Mean Squared Error for tuned model: {mse:.4f}")
print(f"Mean Absolute Error for tuned model: {mae:.4f}")
print(f"R-squared score for tuned model: {r2:.4f}")
````
