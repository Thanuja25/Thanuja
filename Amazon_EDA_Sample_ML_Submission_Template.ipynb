## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Remove duplicate rows
#merge_data=merge_data.drop_duplicates()
#2. Filling missing characters with "not available"
merge_data['character']=merge_data['character'].fillna("not available")
#3. Filling missing seasons with 0
merge_data['seasons']=merge_data['seasons'].fillna(0)
#4. Filling age_certification with "unrated"
merge_data['age_certification']=merge_data['age_certification'].fillna("unrated")
#5. Filling appropriate values
merge_data['imdb_score']=merge_data['imdb_score'].fillna(merge_data['imdb_score'].mean())
merge_data['imdb_votes']=merge_data['imdb_votes'].fillna(0)
merge_data['tmdb_popularity']=merge_data['tmdb_popularity'].fillna(0)
merge_data['tmdb_score']=merge_data['tmdb_score'].fillna(merge_data['tmdb_score'].mean())

#6. data format
import ast

def convert_to_list(x):
    try:
        return ast.literal_eval(x) if isinstance(x, str) else []
    except (SyntaxError, ValueError):
        return []

merge_data['genres'] = merge_data['genres'].apply(convert_to_list)
merge_data['production_countries'] = merge_data['production_countries'].apply(convert_to_list)


#7. Data type conversion
merge_data['release_year'] = merge_data['release_year'].astype(int)
merge_data['runtime']=merge_data['runtime'].astype(int)
merge_data['seasons']=merge_data['seasons'].astype(int)
merge_data['imdb_votes']=merge_data['imdb_votes'].astype(int)
merge_data['tmdb_popularity']=merge_data['tmdb_popularity'].astype(float)

merge_data.info()
merge_data.head()

````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

Null Hypothesis:  
The average IMDB scores for movies and TV shows are same
Alternative Hypothesis:
The average IMDB scores for movies and TV shows are different

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats
#dropping rows with null values in IMDB score
merge_data=merge_data.dropna(subset=['imdb_score'])
#Separate IMDB score for movies and TV shows
movie_score=merge_data[merge_data['type']=='movie']['imdb_score']
tv_score=merge_data[merge_data['type']=='tv show']['imdb_score']
#perform independent t test
t_statistic, p_value= stats.ttest_ind(movie_score,tv_score)
print("hypothesis 1: IMDB score of movies and TV shows" )
print(f"T-statistics:{t_statistic:.4f},p-vlaue:{p_value:.4f}")
if p_value<0.05:
  print("Reject the null hypothesis")
else:
  print("Fail to reject the null hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

Independent T-tes

##### Why did you choose the specific statistical test?

Answer Here.
TO compare mean of two independent group

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

merge_data=merge_data.dropna(subset=['runtime'])
correlation, p_value = stats.pearsonr(merge_data['runtime'],merge_data['imdb_score'])
print("\n hypothesis 2: runtime vs IMDB score correlation")
print(f"person correlation :{correlation:.4f},p-vlaue:{p_value:.4f}")
if p_value<0.05:
  print("Reject the null hypothesis (significant corelation between runtime ans IMDB score)")
else:
  print("Fail to reject the null hypothesis (no significant corelation between runtime and IMDB score)")
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

numeric_columns=['release_year','imdb_score','imdb_votes','tmdb_popularity','tmdb_score']
for column in numeric_columns:
  Q1=merge_data[column].quantile(0.25)
  Q3=merge_data[column].quantile(0.75)
  IQR=Q3-Q1
  Lower_bound=Q1-1.5*IQR
  Upper_bound=Q3+1.5*IQR
  merge_data[column]=merge_data[column].clip(lower=Lower_bound,upper=Upper_bound)

#verify outlier treatment by checking min/max values
outlier_summary=merge_data.describe()
outlier_summary
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Answer Here.

I applied winsorization (capping extreme values) based on the interquartile ranging (IQR)method to prevent to distortion while preserving meanigful variations.
Result after treatment:
No extreme value remain beyond 1.5xIQR

````python
# Manipulate Features to minimize feature correlation and create new features

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#compute correlation matrix
corr_matrix=merge_data.corr()

#visualizing correlation heatmap
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm', linewidth=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()

#Creating new features
current_year=2026
# Transform 'release_year' in-place to 'years since release'
merge_data["release_year"]=current_year-merge_data['release_year']
merge_data["engagement_score"]=merge_data['imdb_score']*merge_data['imdb_votes']

#dropping redundant highly correlated features
# 'release_year' is now a transformed feature, so it should not be dropped if intended to be kept.
drop_cols=[] # Remove 'release_year' from drop_cols if the transformed column is meant to be kept.
merge_data.drop(columns=drop_cols,inplace=True, errors='ignore') # Use errors='ignore' to prevent error if column not found

#verify new features
merge_data.head()

#verify new features
merge_data.head()
````

#### 2. Feature Selection

````python
# Select your features wisely to avoid overfitting
from sklearn.feature_selection import VarianceThreshold, mutual_info_regression

#Step 1: Remove low-variance features
variance_threshold=VarianceThreshold(threshold=0.1)
filtered_data=merge_data.copy()
filtered_data=pd.DataFrame(variance_threshold.fit_transform(filtered_data),columns=merge_data.columns[variance_threshold.get_support()])

#Step 2: Drop highly correlated features
correlation_matrix=filtered_data.corr().abs()
upper_tri=correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape),k=1).astype(bool))
high_corr_cols=[col for col in upper_tri.columns if any(upper_tri[col]>0.9)]
filtered_data.drop(columns=high_corr_cols,inplace=True)

#step 3: select top featues using mutual information
x=filtered_data.drop(columns=['imdb_score'])
y=merge_data['imdb_score']
mi_scores=mutual_info_regression(x,y)
mi_scores=pd.Series(mi_scores,index=x.columns).sort_values(ascending=False)

#step 4: select top 15 most important features
selected_features=mi_scores.head(15).index.tolist()
filtered_data=filtered_data[selected_features+["imdb_score"]]

#verify selected features
filtered_data.head()
````

````python
import logging
import optuna
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# 1. Enable logging so you see progress every few seconds
optuna.logging.set_verbosity(optuna.logging.INFO);

# 2. Use a subset (e.g., 2000 rows) for lightning-fast tuning
# This assumes x_train is a pandas DataFrame
x_train_sub = x_train.sample(n=min(2000, len(x_train)), random_state=42)
y_train_sub = y_train.loc[x_train_sub.index]

def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 10, 50), # Reduced for speed
        'max_depth': trial.suggest_int('max_depth', 2, 5),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 5),
        'bootstrap': True, # Hardcoded to True for faster training
        'n_jobs': -1  # Uses all CPU cores
    }

    model = RandomForestRegressor(**params, random_state=42);

    # Use cv=3 for the subset testing
    score = cross_val_score(model, x_train_sub, y_train_sub, cv=3,
                            scoring='neg_mean_squared_error', n_jobs=-1).mean();
    return score

# 3. Create and run the study
study = optuna.create_study(direction='maximize');
study.optimize(objective, n_trials=10); # 10 trials on 2000 rows will be very fast

# 4. Train the FINAL model on the FULL data using the best params
print(f"Best Params found: {study.best_params}")
best_model = RandomForestRegressor(**study.best_params, random_state=42, n_jobs=-1);
best_model.fit(x_train, y_train); # Full data fit

# Calculate evaluation metrics for the best model
cv_score = cross_val_score(best_model, x_train, y_train, cv=5, scoring='neg_mean_squared_error');
cv_mse = abs(cv_score.mean()); # Using absolute value as scoring='neg_mean_squared_error' returns negative MSE

y_pred = best_model.predict(x_test);
mse = mean_squared_error(y_test, y_pred);
mae = mean_absolute_error(y_test, y_pred);
r2 = r2_score(y_test, y_pred);

print(f"\nCross-validation MSE for tuned model: {cv_mse:.4f}")
print(f"Mean Squared Error for tuned model: {mse:.4f}")
print(f"Mean Absolute Error for tuned model: {mae:.4f}")
print(f"R-squared score for tuned model: {r2:.4f}")
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model

from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import optuna
import numpy as np

def objective(trial):
  params={
     'n_estimators': trial.suggest_int('n_estimators', 10,100),
     'max_depth': trial.suggest_int('max_depth', 2,10),
     'learning_rate': trial.suggest_float('learning_rate', 0.01, 1.0),
     'subsample': trial.suggest_float('subsample', 0.5, 1.0),
     'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5,1.0),
     'gamma': trial.suggest_float('gamma', 0.0, 1.0),
     'reg_alpha': trial.suggest_float('reg_alpha', 1e-8,10.0),
     'reg_lambda': trial.suggest_float('reg_lambda', 1e-8,10.0)
  }
  model=XGBRegressor(**params,random_state=42)
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  return mean_squared_error(y_test,y_pred)

study=optuna.create_study(direction='minimize')
study.optimize(objective,n_trials=10)

best_params=study.best_params
best_xgb_model=XGBRegressor(**best_params,random_state=42)
best_xgb_model.fit(x_train,y_train)

y_pred_xgb_tuned=best_xgb_model.predict(x_test)

mse_xgb_tuned=mean_squared_error(y_test,y_pred_xgb_tuned)
mae_xgb_tuned=mean_absolute_error(y_test,y_pred_xgb_tuned)
R2_xgb_tuned=r2_score(y_test,y_pred_xgb_tuned)

print("XGBoost model after Hyperparameter Tuning")
print(f"mean squared error: {mse_xgb_tuned:.4f}")
print(f"mean absolute error: {mae_xgb_tuned:.4f}")
print(f"R-squared score: {R2_xgb_tuned:.4f}")
````
